### Question 1

```python
import numpy as np

# Define matrix X
X = np.array([[1, 1], [3, 4]])

# Compute the inverse of X
inv_X = np.linalg.inv(X)

# Define vector y
y = np.array([0, 1])

# Compute w = X⁻¹ * y
w = inv_X.dot(y)

# Print the result
print(w)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix X and vector y
m_list = [[1, 1], [3, 4]]
X = np.array(m_list)
y = np.array([0, 1])

# Compute the inverse and solution
inv_X = np.linalg.inv(X)
w = inv_X @ y  # Using @ for matrix multiplication

# Define the range for w1
w1 = np.linspace(-3, 3, 100)

# Solve for w2 in terms of w1 for both equations
eq1 = (y[0] - X[0, 0] * w1) / X[0, 1]  # Equation (1): w1 + w2 = 0
eq2 = (y[1] - X[1, 0] * w1) / X[1, 1]  # Equation (2): 3w1 + 4w2 = 1

# Plot the equations
plt.figure(figsize=(7,5))
plt.plot(w1, eq1, 'r-', label="Equation (1): $w_1 + w_2 = 0$")
plt.plot(w1, eq2, 'b--', label="Equation (2): $3w_1 + 4w_2 = 1$")

# Plot the solution point
plt.scatter(w[0], w[1], color='black', zorder=3, label=f"Solution: (${w[0]:.2f}$, ${w[1]:.2f}$)")

# Formatting the plot
plt.xlabel(r'$w_1$', fontsize=14)
plt.ylabel(r'$w_2$', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
# plt.xlim(-3, 3)
# plt.ylim(-3, 3)

# Show the plot
plt.show()
```

### Question 3

```python
import numpy as np
from numpy.linalg import inv
X = np.array([[1, 2], [2, 4], [1, -1]])
y = np.array([0, 0.1, 1])
w = inv(X.T @ X) @ X.T @ y
print(w)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv

# Define the system
X = np.array([[1, 2], [2, 4], [1, -1]])  # Coefficients of the equations
y = np.array([0, 0.1, 1])  # Output values
w = inv(X.T @ X) @ X.T @ y  # Least squares solution

# Define a range for plotting
w1_vals = np.linspace(-4, 5, 100)

# Compute the lines representing each equation in the system
eq1 = (y[0] - X[0, 0] * w1_vals) / X[0, 1]
eq2 = (y[1] - X[1, 0] * w1_vals) / X[1, 1]
eq3 = (y[2] - X[2, 0] * w1_vals) / X[2, 1]

# Plot the equations
plt.figure(figsize=(8, 6))
plt.plot(w1_vals, eq1, 'r-', label='Equation (1)')  # Red solid line
plt.plot(w1_vals, eq2, 'b--', label='Equation (2)')  # Blue dashed line
plt.plot(w1_vals, eq3, 'k:', label='Equation (3)')  # Black dotted line

# Scatter the approximated w solution
plt.scatter(w[0], w[1], color='black', marker='o', label='Approximated w')

# Labels and Legend
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.legend()
plt.title("Overdetermined System and Least Squares Solution")
plt.grid(True)

plt.xlim(-2, 4)
plt.ylim(-1.5, 1)

# Show plot
plt.show()
```

```python
import numpy as np

# Define the matrix X and vector y
X = np.array([[1, 0, 1, 0], 
              [1, -1, 1, -1], 
              [1, 1, 0, 0]])

y = np.array([1, 0, 1])

# Compute the Moore-Penrose pseudo-inverse solution
w_hat = X.T @ np.linalg.inv(X @ X.T) @ y

# Print the computed w_hat
print("w_hat:", w_hat)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the matrix X and vector y
X = np.array([[1, 2, 1], [2, 4, -1]])
y = np.array([0, 1])

# Define a grid for visualization
w3_range = np.linspace(-2, 2, 30)  # X-axis (w3, rightmost)
w2_range = np.linspace(-2, 2, 30)  # Y-axis (w2, correct direction)
W3, W2 = np.meshgrid(w3_range, w2_range)

# Compute the corresponding values of w1 for both equations
W1_eq1 = (y[0] - X[0, 1] * W2 - X[0, 2] * W3) / X[0, 0]  # Equation 1
W1_eq2 = (y[1] - X[1, 1] * W2 - X[1, 2] * W3) / X[1, 0]  # Equation 2

# Create 3D figure
fig = plt.figure(figsize=(13,13))
ax = fig.add_subplot(111, projection='3d')

# Plot the two planes as wireframes with color gradient
ax.plot_wireframe(W3, W2, W1_eq1, color='gold', alpha=0.7)  # Keep w2 direction unchanged
ax.plot_wireframe(W3, W2, W1_eq2, color='blue', alpha=0.7)

# Labels and title
ax.set_xlabel(r'$w_3$', fontsize=14) 
ax.set_ylabel(r'$w_2$', fontsize=14)  
ax.set_zlabel(r'$w_1$', fontsize=14) 

# Adjust axis directions to match expected layout
ax.invert_yaxis()  # Flip w2 axis instead of modifying function

# Adjust the view angle so that w1 (Z-axis) is on the left
ax.view_init(elev=10, azim=210)  # Rotate axes so Z is on the left

# Show the plot
plt.show()
```

```python
import numpy as np

# Define the matrix A
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# (a) Check if A is invertible (only square matrices can be invertible)
is_invertible = A.shape[0] == A.shape[1] and np.linalg.det(A) != 0 if A.shape[0] == A.shape[1] else False

# (b) Check if A is left invertible (A^T A should be invertible)
ATA = A.T @ A
det_ATA = np.linalg.det(ATA) if np.linalg.matrix_rank(ATA) == min(A.shape) else 0
is_left_invertible = det_ATA != 0

# (c) Check if A is right invertible (A A^T should be invertible)
AAT = A @ A.T
det_AAT = np.linalg.det(AAT) if np.linalg.matrix_rank(AAT) == min(A.shape) else 0
is_right_invertible = det_AAT != 0

# (d) Check if A has a determinant (only square matrices have determinants)
has_determinant = A.shape[0] == A.shape[1]

# Print results
results = {
    "(a) A is invertible": is_invertible,
    "(b) A is left invertible": is_left_invertible,
    "(c) A is right invertible": is_right_invertible,
    "(d) A has no determinant": not has_determinant
}

results
```
